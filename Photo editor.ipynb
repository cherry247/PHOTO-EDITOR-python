```python
# importing libraries
from tkinter import *
from PIL import ImageTk,Image
import cv2
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,NavigationToolbar2Tk) 
import matplotlib.image as mpimg
```

```python
#create a class that contains your buttons and various filters
class testme:
    def __init__(self,frame1):
        self.frame1=frame1
        #self.frame1.pack()
        
        self.button=Button(self.frame1,text="original",command=self.upload) 
       
        self.button.pack(fill=BOTH,side=LEFT)
        
        self.button1=Button(self.frame1,text="reset",command=self.clearme)
        self.button1.pack(fill=BOTH,side=LEFT)
       
        self.button2=Button(self.frame1,text=" Spectral",command=self.filter1)
        self.button2.pack(fill=BOTH,side=LEFT)
        
        self.button3=Button(self.frame1,text="summer",command=self.filter2)
        self.button3.pack(fill=BOTH,side=LEFT)
        
        self.button4=Button(self.frame1,text="twilight",command=self.filter3)
        self.button4.pack(fill=BOTH,side=LEFT)
        
        self.button5=Button(self.frame1,text="pastel",command=self.filter4)
        self.button5.pack(fill=BOTH,side=LEFT)
       
           

    def upload(self):                   
        fig = Figure(figsize = (3,3))
        img_arr = mpimg.imread('CarryMinati.png')
        f = Figure()
        a = f.add_subplot()
        a.imshow(img_arr)
        a.axis('off')
        self.wierdobject = FigureCanvasTkAgg(f, master=self.frame1) 
        self.wierdobject.get_tk_widget().pack() 
        self.wierdobject.draw()     
        
    def filter1(self):  
        fig = Figure(figsize = (3,3))
        img_arr = mpimg.imread('CarryMinati.png')
        f = Figure()
        a = f.add_subplot()
        a.imshow(img_arr[:,:,2],cmap='Spectral')
        a.axis('off')
        self.wierdobject = FigureCanvasTkAgg(f, master=self.frame1) 
        self.wierdobject.get_tk_widget().pack() 
        self.wierdobject.draw()  
    def filter2(self):  
        fig = Figure(figsize = (3,3))
        img_arr = mpimg.imread('CarryMinati.png')
        f = Figure()
        a = f.add_subplot()
        a.imshow(img_arr[:,:,2])
        a.axis('off')
        self.wierdobject = FigureCanvasTkAgg(f, master=self.frame1) 
        self.wierdobject.get_tk_widget().pack() 
        self.wierdobject.draw()
    def filter3(self):  
        fig = Figure(figsize = (3,3))
        img_arr = mpimg.imread('CarryMinati.png')
        f = Figure()
        a = f.add_subplot()
        a.imshow(img_arr[:,:,0],cmap='twilight')
        a.axis('off')
        self.wierdobject = FigureCanvasTkAgg(f, master=self.frame1) 
        self.wierdobject.get_tk_widget().pack() 
        self.wierdobject.draw()
    def filter4(self):  
        fig = Figure(figsize = (3,3))
        img_arr = mpimg.imread('CarryMinati.png')
        f = Figure()
        a = f.add_subplot()
        a.imshow(img_arr[:,:,0],cmap='Pastel1')
        a.axis('off')
        self.wierdobject = FigureCanvasTkAgg(f, master=self.frame1) 
        self.wierdobject.get_tk_widget().pack() 
        self.wierdobject.draw()
    
       

    def clearme(self):       
        self.wierdobject.get_tk_widget().pack_forget()     
```

```python
# running this cell will open a gui window
#make sure to reset everytime when you switch different filters
root=Tk()
root.title('PHOTO EDITOR')
root.geometry('500x200')
aframe=Frame(root)
testme(aframe)
aframe.pack()  #packs a frame which given testme packs frame 1 in testme
root.mainloop()
```
